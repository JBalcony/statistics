# Przykład 
Zmienna losowa o rozkładzie Normalnym


```python
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

# Zad 1.
Policzmy pierwsze momenty dla zmiennej losowej o rozkładzie normalnym

```python
mean, var, skew, kurt = norm.stats(moments='mvsk')
print(mean)
print(var)
print(skew)
print(kurt)
```

# Zad 2.
Narysujmy gęstość zmiennej losowej o rozkładzie normalnym

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x), 'r-', lw=5, alpha=0.6, label='norm pdf')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
rv = norm()
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
plt.show()
```

# Zad 3
Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
r = norm.rvs(size=1000)
fig, ax = plt.subplots(1, 1)
ax.hist(r, normed=True, histtype='stepfilled', alpha=0.2, label='histogram')
ax.legend(loc='best', frameon=False)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='m=0, sigma=0')
ax.legend()
plt.show()
```

# Zad 4
Narysuj dystrybuantę rozkładu Normalnego

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, norm.cdf(x, 0, 1), '--', ms=8, label='bernoulli cdf')
plt.show()
```

# Zad 5
Narysuj gęstości rozkładu normalnego z różnymi parametrami

```python
#------------------------------------------------------------
# Define the distributions to be plotted
sigma_values = [0.5, 1.0, 2.0]
mu_values = [0, 1, 2]
#mu = 0
x = np.linspace(-10, 10, 1000)

#------------------------------------------------------------
# plot the distributions
fig, ax = plt.subplots(figsize=(5, 3.75))

for sigma, mu in zip(sigma_values, mu_values):
    # create a gaussian / normal distribution
    dist = norm(mu, sigma)
    plt.plot(x, dist.pdf(x), ls="--", c='black', label=r'$\mu=%1.f,\ \sigma=%.1f$' % (mu, sigma))
plt.legend()
plt.show()
```

# Zad 6
Narysuj kilka histogramów różnych próbek z roskładu normalnego o tych samych parametrach.

```python
#------------------------------------------------------------
# Define the distributions to be plotted
sigma_values = [0.5, 1.0, 2.0]
mu_values = [0, 1, 2]

fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        r = norm.rvs(size=100)
        #ax[i].plot(x,x)
        ax[i,j].hist(r, normed=True)
plt.show()
```

# Zad 7
Zwizuaizuj zasadę 3 sigm

```python
def show_SD(axis, xi, yi, text):
    '''Show the area covered by 1/2/3 SDs'''
    
    fc = '#DDDDDD'
    axis.plot(x,y)
    axis.fill_between(xi, yi, facecolor=fc)
    
    axis.text(0, 0.05, text, horizontalalignment='center', fontsize=25)
    axis.set_xlim([-3.5, 3.5])
    axis.set_ylim([-0.0, 0.5])
    axis.set_yticks([])
```

```python
nd = norm()
x = np.linspace(-3.5, 3.5,100)
x1 = np.linspace(-1,1)
x2 = np.linspace(-2,2)
x3 = np.linspace(-3,3)

y = nd.pdf(x)
y1 = nd.pdf(x1)
y2 = nd.pdf(x2)
y3 = nd.pdf(x3)

```

```python
fig, axs = plt.subplots(1,3, sharey=True, figsize=(10, 7))
show_SD(axs[0], x1, y1, '68.3%')    
show_SD(axs[1], x2, y2, '95.4%')    
show_SD(axs[2], x3, y3, '99.7%')  
plt.show()
```

```python
mu = -2
sigma = 0.7
myDistribution = stats.norm(mu, sigma)
significanceLevel = 0.05
myDistribution.ppf([significanceLevel/2, 1-significanceLevel/2] )
```

# Zad 8
Narysuj funkcję PPF (odwrotna do dystrybuanty CDF) rozkładu Normalnego.

```python
# 100 numbers between 0 and 1.0 i.e., probabilities.
x = np.linspace(0, 1.0, 100)
# PPF at these values
y = norm.ppf(x, loc=0, scale=1)

plt.plot(x,y, color="black")
plt.xlabel("Cumulative Probability")
plt.ylabel("Variate")

plt.title("PPF for Gaussian of mean = {0} & std. deviation = {1}".format(0, 1))
plt.show()
```
