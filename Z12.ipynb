```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

# Rozkład t-Studenta

```python
t = np.arange(-5, 5, 0.05)
TVals = [1,5]
```

## Gęstość

```python
normal = stats.norm.pdf(t)
t1 = stats.t.pdf(t,1)
t5 = stats.t.pdf(t,5)

plt.plot(t,normal, '--',  label='normal')
plt.plot(t, t1, label='df=1')
plt.plot(t, t5, label='df=5')
plt.legend()

plt.xlim(-5,5)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.axis('tight')
plt.show()
```

## Dystrybuanta

```python
normal = stats.norm.cdf(t)
t1 = stats.t.cdf(t,1)
t5 = stats.t.cdf(t,5)

plt.plot(t,normal, '--',  label='normal')
plt.plot(t, t1, label='df=1')
plt.plot(t, t5, label='df=5')
plt.legend()

plt.xlim(-5,5)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.axis('tight')
plt.show()
```

# chi-kwadrat

```python
t = np.arange(0, 8, 0.05)
Chi2Vals = [1,2,3,5]
```

## Gęstość

```python
for chi2 in Chi2Vals:
    plt.plot(t, stats.chi2.pdf(t, chi2), label='k={0}'.format(chi2))
plt.legend()

plt.xlim(0,8)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.axis('tight')

plt.show()
```

# Dystrybuanta

```python
t = np.arange(-1, 5, 0.05)
for chi2 in Chi2Vals:
    plt.plot(t, stats.chi2.cdf(t, chi2), label='k={0}'.format(chi2))
plt.legend()

plt.xlim(0,8)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.axis('tight')

plt.show()
```

# F-distribution

```python
t = np.arange(0, 3, 0.01)
d1s = [1,2,5,100]
d2s = [1,1,2,100]
```

## Gęstość

```python
for (d1,d2) in zip(d1s,d2s):
    plt.plot(t, stats.f.pdf(t, d1, d2), label='F({0}/{1})'.format(d1,d2))
plt.legend()

plt.xlim(0,3)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.axis('tight')
plt.legend()
plt.show()
```

## Dystrybuanta

```python
for (d1,d2) in zip(d1s,d2s):
    plt.plot(t, stats.f.cdf(t, d1, d2), label='F({0}/{1})'.format(d1,d2))
plt.legend()

plt.xlim(0,3)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.axis('tight')
plt.legend()
plt.show()
```

# Rozkład wykładniczy

```python
t = np.arange(0, 3, 0.01)
lambdas = [0.5, 1, 1.5]
```

## Gęstość

```python
for par in lambdas:
    plt.plot(t, stats.expon.pdf(t, 0, par), label='$\lambda={0:3.1f}$'.format(par))
plt.legend()

plt.xlim(0,3)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.axis('tight')
plt.legend()
plt.show()
```

## Dystrybuanta

```python
for par in lambdas:
    plt.plot(t, stats.expon.cdf(t, 0, par), label='$\lambda={0:3.1f}$'.format(par))
plt.legend()

plt.xlim(0,3)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.axis('tight')
plt.legend()
plt.show()
```

# Rozkład Weibulla

```python
t = np.arange(0, 2.5, 0.01)
lambdaVal = 1
ks = [0.5, 1, 1.5, 5]
```

## Gęstość

```python
for k in ks:
    wd = stats.weibull_min(k)
    plt.plot(t, wd.pdf(t), label='k = {0:.1f}'.format(k))

plt.xlim(0,2.5)
plt.ylim(0,2.5)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.legend()
plt.show()
```

## Dystrybuanta

```python
for k in ks:
    wd = stats.weibull_min(k)
    plt.plot(t, wd.cdf(t), label='k = {0:.1f}'.format(k))

plt.xlim(0,2.5)
plt.ylim(0,2.5)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.legend()
plt.show()
```

## Rozkład logarytmicznie normalny

```python
t = np.arange(0, 2.5, 0.01)
s = [0.5, 1, 1.5, 5]
```

## Gęstość

```python
for k in s:
    wd = stats.lognorm(k)
    plt.plot(t, wd.pdf(t), label='k = {0:.1f}'.format(k))

plt.xlim(0,2.5)
plt.ylim(0,2.5)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.legend()
plt.show()
```

## Dystrybuanta

```python
for k in s:
    wd = stats.lognorm(k)
    plt.plot(t, wd.cdf(t), label='k = {0:.1f}'.format(k))

plt.xlim(0,2.5)
plt.ylim(0,2.5)
plt.xlabel('X')
plt.ylabel('pdf(X)')
plt.legend()
plt.show()
```
